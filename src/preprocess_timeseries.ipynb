# Preprocess Dataset

Notebook to select the CPI index attribute and convert the data to a normal pandas dataframe to be digested

```python
import pandas as pd
import os
```

```python
metadata = pd.read_csv(os.path.join('..', 'CPITimeSeries', 'Metadata_CPI_09-22-2022 10-43-47-47_timeSeries.csv'))
metadata
```

```python
data = pd.read_csv(os.path.join('..', 'CPITimeSeries', 'CPITimeSeries.csv'))
data.columns
```

```python
indicator = 'PCPI_IX'
```

```python
indicator_data = pd.DataFrame(data.loc[(data['Indicator Code'] == indicator) & (data.Attribute == 'Value')])
```

```python
indicator_data = indicator_data.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Attribute', 'Common Reference Period', 'Unnamed: 878'], axis=1)
```

```python
indicator_data
```

```python
indicator_data = indicator_data.T
```

```python
indicator_data
```

```python
indicator_data.columns = indicator_data.iloc[0]
indicator_data = indicator_data[1:]
```

```python
indicator_data
```

```python
indicator_data
```

```python
time = pd.Series(pd.date_range(start='1950/01/01', end='2022/08/01', freq='MS'), name='Date')
time
```

```python
indicator_data.index = time
```

```python
indicator_data
```

```python
indicator_data.columns
```

```python
for col in indicator_data.columns:
    indicator_data[col] = indicator_data[col].astype(float)
```

```python
indicator_data
```

```python
indicator_data.to_csv(os.path.join('..', 'CPITimeSeries', 'time_series_data.csv'))
```
